# Phần 2 — Trực quan hoá & Phân tích Dữ liệu (Data Storytelling)

```python
import warnings
warnings.filterwarnings("ignore")

from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid", context="talk")
plt.rcParams["figure.figsize"] = (13, 6)
plt.rcParams["axes.titlesize"] = 16
plt.rcParams["axes.labelsize"] = 12

DATA_DIR = Path(r"c:/Users/ADMIN/Documents/Datathon/datathon-2026-round-1")
```

```python
# Load all CSVs
customers = pd.read_csv(DATA_DIR / "customers.csv")
geography = pd.read_csv(DATA_DIR / "geography.csv")
inventory = pd.read_csv(DATA_DIR / "inventory.csv")
order_items = pd.read_csv(DATA_DIR / "order_items.csv", low_memory=False)
orders = pd.read_csv(DATA_DIR / "orders.csv")
payments = pd.read_csv(DATA_DIR / "payments.csv")
products = pd.read_csv(DATA_DIR / "products.csv")
promotions = pd.read_csv(DATA_DIR / "promotions.csv")
returns = pd.read_csv(DATA_DIR / "returns.csv")
reviews = pd.read_csv(DATA_DIR / "reviews.csv")
sales = pd.read_csv(DATA_DIR / "sales.csv")
sample_submission = pd.read_csv(DATA_DIR / "sample_submission.csv")
shipments = pd.read_csv(DATA_DIR / "shipments.csv")
web_traffic = pd.read_csv(DATA_DIR / "web_traffic.csv")

# Parse date columns
date_cols = {
    "orders": ["order_date"],
    "customers": ["signup_date"],
    "returns": ["return_date"],
    "reviews": ["review_date"],
    "shipments": ["ship_date", "delivery_date"],
    "promotions": ["start_date", "end_date"],
    "sales": ["Date"],
    "sample_submission": ["Date"],
    "web_traffic": ["date"],
    "inventory": ["snapshot_date"],
}

for c in date_cols["orders"]:
    orders[c] = pd.to_datetime(orders[c], errors="coerce")
for c in date_cols["customers"]:
    customers[c] = pd.to_datetime(customers[c], errors="coerce")
for c in date_cols["returns"]:
    returns[c] = pd.to_datetime(returns[c], errors="coerce")
for c in date_cols["reviews"]:
    reviews[c] = pd.to_datetime(reviews[c], errors="coerce")
for c in date_cols["shipments"]:
    shipments[c] = pd.to_datetime(shipments[c], errors="coerce")
for c in date_cols["promotions"]:
    promotions[c] = pd.to_datetime(promotions[c], errors="coerce")
for c in date_cols["sales"]:
    sales[c] = pd.to_datetime(sales[c], errors="coerce")
for c in date_cols["sample_submission"]:
    sample_submission[c] = pd.to_datetime(sample_submission[c], errors="coerce")
for c in date_cols["web_traffic"]:
    web_traffic[c] = pd.to_datetime(web_traffic[c], errors="coerce")
for c in date_cols["inventory"]:
    inventory[c] = pd.to_datetime(inventory[c], errors="coerce")

all_data = {
    "customers": customers,
    "geography": geography,
    "inventory": inventory,
    "order_items": order_items,
    "orders": orders,
    "payments": payments,
    "products": products,
    "promotions": promotions,
    "returns": returns,
    "reviews": reviews,
    "sales": sales,
    "sample_submission": sample_submission,
    "shipments": shipments,
    "web_traffic": web_traffic,
}

pd.DataFrame({
    "dataset": list(all_data.keys()),
    "rows": [df.shape[0] for df in all_data.values()],
    "cols": [df.shape[1] for df in all_data.values()],
}).sort_values("rows", ascending=False)
```

### 1. Xử lí missing và outlier

```python
print("="*80)
print("CUSTOMERS - Missing Data")
print("="*80)
missing_cust = customers.isna().sum().sort_values(ascending=False)
missing_pct_cust = (missing_cust / len(customers) * 100).round(2)
result_cust = pd.DataFrame({
    "Column": missing_cust.index,
    "Missing_Count": missing_cust.values,
    "Missing_%": missing_pct_cust.values,
})
print(result_cust.to_string(index=False))
```

```python
print("="*80)
print("GEOGRAPHY - Missing Data")
print("="*80)

missing_geo = geography.isna().sum().sort_values(ascending=False)
missing_pct_geo = (missing_geo / len(geography) * 100).round(2)

result_geo = pd.DataFrame({
    "Column": missing_geo.index,
    "Missing_Count": missing_geo.values,
    "Missing_%": missing_pct_geo.values,
})

print(result_geo.to_string(index=False))
```

```python
print("="*80)
print("ORDERS - Missing Data")
print("="*80)

missing_orders = orders.isna().sum().sort_values(ascending=False)
missing_pct_orders = (missing_orders / len(orders) * 100).round(2)

result_orders = pd.DataFrame({
    "Column": missing_orders.index,
    "Missing_Count": missing_orders.values,
    "Missing_%": missing_pct_orders.values,
})

print(result_orders.to_string(index=False))
```

```python
print("="*80)
print("PRODUCTS - Missing Data")
print("="*80)

missing_products = products.isna().sum().sort_values(ascending=False)
missing_pct_products = (missing_products / len(products) * 100).round(2)

result_products = pd.DataFrame({
    "Column": missing_products.index,
    "Missing_Count": missing_products.values,
    "Missing_%": missing_pct_products.values,
})

print(result_products.to_string(index=False))


print("\n" + "="*80)
print("PRODUCTS - Outlier (price & cogs)")
print("="*80)

cols_to_check = ["price", "cogs"]

for col in cols_to_check:
    if col in products.columns:
        Q1 = products[col].quantile(0.25)
        Q3 = products[col].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = products[(products[col] < lower_bound) | (products[col] > upper_bound)]

        if len(outliers) > 0:
            print(f"  {col}: {len(outliers):,} outliers (bounds: {lower_bound:.2f} - {upper_bound:.2f})")
        else:
            print(f"  {col}: No outliers")
    else:
        print(f"  {col}: Column not found")
```

```python
print("="*80)
print("ORDER_ITEMS - Missing Data")
print("="*80)

missing_oi = order_items.isna().sum().sort_values(ascending=False)
missing_pct_oi = (missing_oi / len(order_items) * 100).round(2)

result_oi = pd.DataFrame({
    "Column": missing_oi.index,
    "Missing_Count": missing_oi.values,
    "Missing_%": missing_pct_oi.values,
})

print(result_oi.to_string(index=False))


print("\n" + "="*80)
print("ORDER_ITEMS - Outlier")
print("="*80)

cols_to_check = ["discount_amount", "unit_price"]

for col in cols_to_check:
    if col in order_items.columns:
        Q1 = order_items[col].quantile(0.25)
        Q3 = order_items[col].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = order_items[
            (order_items[col] < lower_bound) | (order_items[col] > upper_bound)
        ]

        if len(outliers) > 0:
            print(f"  {col}: {len(outliers):,} outliers (bounds: {lower_bound:.2f} - {upper_bound:.2f})")
        else:
            print(f"  {col}: No outliers")
    else:
        print(f"  {col}: Column not found")
```

### 1.1 CUSTOMERS & GEOGRAPHY — Nhận xét

```python
print("🔴 PRODUCTS — CRITICAL FINDINGS: Ghost Twins & Ultra-Cheap SKUs")
print("="*80)

# Step 1: Identify ultra-cheap products (likely unit error)
ultra_cheap = products[products["price"] < 100]
print(f"\n🔴 Ultra-cheap products (< 100 VND): {len(ultra_cheap):,} SKUs")
print(f"   → Có thể là lỗi đơn vị: VND vs nghìn VND")

if len(ultra_cheap) > 0:
    print(f"   Examples:")
    for idx, row in ultra_cheap.head(5).iterrows():
        print(f"     • {row['product_name']} ({row['category']}) - giá: {row['price']:.0f} VND")

# Step 2: Join với order_items để check nếu thực tế có ai mua
ultra_cheap_sold = order_items.merge(ultra_cheap[["product_id"]], on="product_id", how="inner")
print(f"\n   Các ultra-cheap trong order_items: {len(ultra_cheap_sold):,} line items")
print(f"   → Số unique ultra-cheap SKU bán được: {ultra_cheap_sold['product_id'].nunique()}")

if ultra_cheap_sold['product_id'].nunique() == 0:
    print(f"   ✓ GOOD NEWS: Không SKU ultra-cheap nào được mua → an toàn để xóa")
else:
    print(f"   ⚠ WARNING: {ultra_cheap_sold['product_id'].nunique()} ultra-cheap SKU đã bán → cần cẩn thận")

# Step 3: Price variance analysis
print(f"\n📊 Price Distribution:")
print(f"   Min price: {products['price'].min():10.0f} VND")
print(f"   25th %ile: {products['price'].quantile(0.25):10.0f} VND")
print(f"   Median: {products['price'].quantile(0.50):10.0f} VND")
print(f"   75th %ile: {products['price'].quantile(0.75):10.0f} VND")
print(f"   Max price: {products['price'].max():10.0f} VND")
print(f"   Ratio (max/min): {products['price'].max() / (products['price'].min() + 1e-6):.0f}x")

# Step 4: COGS vs Price check
products["margin"] = products["price"] - products["cogs"]
negative_margin = (products["margin"] < 0).sum()
print(f"\n💰 Margin Check:")
print(f"   Negative margin (price < COGS): {negative_margin:,} SKUs")
if negative_margin > 0:
    print(f"   ⚠ DATA ERROR: {negative_margin} products would lose money at list price")

print("\n✅ RECOMMEND:")
print(f"   1. 🗑️  DROP {len(ultra_cheap)} ultra-cheap SKU (< 100 VND) → chưa bao giờ được mua")
print(f"   2. 🔍 INVESTIGATE price extremes → may be unit encoding error")
print(f"   3. 📍 FLAG {negative_margin} negative margin records → impossible pricing")
```

```python
print("🔴 SALES — CRITICAL FINDINGS: Loss Days & Margin Compression")
print("="*80)

# Step 1: Calculate profit margin
sales_copy = sales.copy()
sales_copy["Profit"] = sales_copy["Revenue"] - sales_copy["COGS"]
sales_copy["Margin_Pct"] = 100 * sales_copy["Profit"] / (sales_copy["Revenue"] + 1e-6)

loss_days = sales_copy[sales_copy["Margin_Pct"] < 0]
total_days = len(sales_copy)

print(f"\n🔴 Loss Day Statistics:")
print(f"   Total records: {total_days:,}")
print(f"   Loss days (margin < 0): {len(loss_days):,} ({100*len(loss_days)/total_days:.1f}%)")

if len(loss_days) > 0:
    print(f"   Avg margin loss: {loss_days['Margin_Pct'].mean():.2f}%")
    print(f"   Worst margin: {loss_days['Margin_Pct'].min():.2f}%")

# Step 2: Temporal patterns
sales_copy["Date"] = pd.to_datetime(sales_copy["Date"])
sales_copy["month"] = sales_copy["Date"].dt.month
sales_copy["year"] = sales_copy["Date"].dt.year

loss_by_month = sales_copy.groupby("month").agg({
    "Margin_Pct": ["count", lambda x: (x < 0).sum(), lambda x: 100*(x<0).sum()/len(x) if len(x)>0 else 0],
}).reset_index()
loss_by_month.columns = ["month", "total_days", "loss_days", "loss_pct"]

print(f"\n📅 Loss by Month (seasonal):")
for _, row in loss_by_month.sort_values("loss_pct", ascending=False).head(5).iterrows():
    print(f"   Month {int(row['month']):2d}: {int(row['loss_pct']):3.0f}% loss ({int(row['loss_days']):3.0f}/{int(row['total_days']):3.0f} days)")

# Step 3: Revenue & cost impact
print(f"\n💰 Financial Impact:")
print(f"   Total revenue (all days): {sales['Revenue'].sum():,.0f} VND")
print(f"   Total cost (all days): {sales['COGS'].sum():,.0f} VND")
if len(loss_days) > 0:
    print(f"   Revenue from loss days: {loss_days['Revenue'].sum():,.0f} VND")
    print(f"   Cost from loss days: {loss_days['COGS'].sum():,.0f} VND")
    print(f"   Gross loss amount: {loss_days['COGS'].sum() - loss_days['Revenue'].sum():,.0f} VND")
    print(f"   → Equivalent to {(loss_days['COGS'].sum() - loss_days['Revenue'].sum()) / sales['Revenue'].sum() * 100:.2f}% of total revenue")

# Step 4: Margin distribution
print(f"\n📊 Margin Distribution:")
print(f"   Margin > 30%: {(sales_copy['Margin_Pct'] > 30).sum():,} days ({100*(sales_copy['Margin_Pct'] > 30).sum()/len(sales_copy):.1f}%)")
print(f"   Margin 10–30%: {((sales_copy['Margin_Pct'] >= 10) & (sales_copy['Margin_Pct'] <= 30)).sum():,} days")
print(f"   Margin 0–10%: {((sales_copy['Margin_Pct'] > 0) & (sales_copy['Margin_Pct'] < 10)).sum():,} days")
print(f"   Margin < 0%: {(sales_copy['Margin_Pct'] < 0).sum():,} days ({100*(sales_copy['Margin_Pct'] < 0).sum()/len(sales_copy):.1f}%)")

print("\n✅ VERDICT: 🔴 CRITICAL")
print("   • 10% loss days = likely flash sales (margin sacrifice for volume)")
print("   • Aug/Dec peaks = seasonal promotional pressure")
print("   • Recommended ACTION:")
print("     1. 📍 FLAG loss days but KEEP (business decision, not data error)")
print("     2. 📊 ANALYZE margin trend → is it intentional strategy or process failure?")
print("     3. 🎯 SEGMENT out flash sale days for separate funnel/ROI analysis")
```

```python
print("🔴 INVENTORY — CRITICAL FINDINGS: Paradox Flags & Dead Stock")
print("="*80)

# Step 1: Paradox analysis (both stockout AND overstock TRUE)
both_flags = (inventory["stockout_flag"] == 1) & (inventory["overstock_flag"] == 1)
paradox_count = both_flags.sum()
paradox_pct = 100 * paradox_count / len(inventory)

print(f"\n🔴 Inventory Status Paradox:")
print(f"   Both stockout & overstock TRUE: {paradox_count:,} ({paradox_pct:.1f}%)")

stockout_only = (inventory["stockout_flag"] == 1) & (inventory["overstock_flag"] == 0)
overstock_only = (inventory["stockout_flag"] == 0) & (inventory["overstock_flag"] == 1)
normal = (inventory["stockout_flag"] == 0) & (inventory["overstock_flag"] == 0)

print(f"   Stockout only: {stockout_only.sum():,}")
print(f"   Overstock only: {overstock_only.sum():,}")
print(f"   Normal (neither): {normal.sum():,}")

# Step 2: Dead stock analysis (high days_of_supply)
print(f"\n📦 Days of Supply (inventory age proxy):")
for pctl in [50, 75, 90, 95, 99]:
    val = inventory["days_of_supply"].quantile(pctl/100)
    print(f"   {pctl}th percentile: {val:.0f} days")

dead_stock_threshold = 180  # ~6 months
dead_stock = inventory[inventory["days_of_supply"] > dead_stock_threshold]
print(f"\n💀 Dead Stock (> {dead_stock_threshold} days):")
print(f"   Count: {len(dead_stock):,} ({100*len(dead_stock)/len(inventory):.1f}%)")
print(f"   Total quantity: {dead_stock['stock_on_hand'].sum():,} units")

# Step 3: Relationship between flags and days_of_supply
print(f"\n📊 Relationship: Flags vs. Days of Supply:")
for label, mask in [
    ("Stockout+Overstock", both_flags),
    ("Stockout only", stockout_only),
    ("Overstock only", overstock_only),
    ("Normal", normal),
]:
    subset = inventory.loc[mask]
    if len(subset) > 0:
        avg_dos = subset["days_of_supply"].mean()
        pctl_95 = subset["days_of_supply"].quantile(0.95)
        print(f"   {label:20s}: avg={avg_dos:7.0f} days, p95={pctl_95:7.0f} days, count={len(subset):,}")

# Step 4: Check if paradox is measurement artifact
if paradox_count > 0:
    paradox_subset = inventory.loc[both_flags]
    print(f"\n🔍 Paradox root cause investigation:")
    print(f"   Avg days_of_supply (paradox): {paradox_subset['days_of_supply'].mean():.0f} days")
    print(f"   → Likely cause: high days_of_supply → both low sales velocity AND high threshold")
    print(f"   → Could be flag logic error: (daily_sales < threshold) AND (qty_on_hand > threshold) = TRUE for BOTH")

print("\n✅ VERDICT: 🔴 CRITICAL")
print("   • {:.1f}% paradox flag = likely measurement/threshold logic error".format(paradox_pct))
print("   • {:.0f}% dead stock (>{} days) = serious working capital issue".format(100*len(dead_stock)/len(inventory), dead_stock_threshold))
print("   • ACTION:")
print("     1. 🔧 INVESTIGATE flag logic → separate stockout/overstock into thresholds")
print("     2. 📍 FLAG paradox records + validate manually")
print("     3. 💰 ANALYZE dead stock SKUs → clearance/discontinuation candidates")
print("     4. 🔄 Consider segmenting inventory age by product lifecycle stage")
```

```python
print("🟡 ORDERS — WARNING: Stuck Orders (Created/Paid Status)")
print("="*80)

# Step 1: Order status distribution
status_counts = orders["order_status"].value_counts()
print(f"\n📊 Order Status Distribution:")
for status, count in status_counts.items():
    pct = 100 * count / len(orders)
    print(f"   {status:15s}: {count:7,} ({pct:5.1f}%)")

# Step 2: Stuck order definition (created or paid, but never progressed)
stuck_mask = orders["order_status"].isin(["created", "paid"])
stuck_orders = orders[stuck_mask]
print(f"\n🔴 Stuck Orders (status = 'created' or 'paid'):")
print(f"   Count: {len(stuck_orders):,} ({100*len(stuck_orders)/len(orders):.1f}%)")

# Step 3: Age of stuck orders
stuck_orders_with_date = stuck_orders.copy()
stuck_orders_with_date["order_date"] = pd.to_datetime(stuck_orders_with_date["order_date"])
stuck_orders_with_date["age_days"] = (pd.Timestamp.now() - stuck_orders_with_date["order_date"]).dt.days

print(f"\n⏰ Age of Stuck Orders (age since creation):")
print(f"   Min age: {stuck_orders_with_date['age_days'].min()} days")
print(f"   Max age: {stuck_orders_with_date['age_days'].max()} days")
print(f"   Median age: {stuck_orders_with_date['age_days'].median():.0f} days")
print(f"   Avg age: {stuck_orders_with_date['age_days'].mean():.0f} days")

# Step 4: Temporal pattern of stuck orders
if "order_date" in orders.columns:
    orders_with_date = orders.copy()
    orders_with_date["order_date"] = pd.to_datetime(orders_with_date["order_date"])
    orders_with_date["year"] = orders_with_date["order_date"].dt.year
    stuck_by_year = orders_with_date[stuck_mask].groupby("year").size()
    print(f"\n📅 Stuck Orders by Creation Year:")
    for year, count in stuck_by_year.items():
        total_that_year = len(orders_with_date[orders_with_date["year"] == year])
        pct = 100 * count / total_that_year
        print(f"   {int(year)}: {count:5,} stuck ({pct:5.1f}% of {total_that_year:,} total)")

# Step 5: Revenue impact
stuck_revenue = stuck_orders["order_total"].sum() if "order_total" in stuck_orders.columns else 0
if stuck_revenue > 0:
    total_revenue = orders["order_total"].sum() if "order_total" in orders.columns else 0
    print(f"\n💰 Revenue at Risk:")
    print(f"   Stuck order total: {stuck_revenue:,.0f} VND")
    print(f"   % of all orders: {100*stuck_revenue/total_revenue:.1f}%")
else:
    print(f"\n💰 No order_total column; can't assess revenue impact")

print("\n✅ VERDICT: 🟡 WARNING (not critical, but needs investigation)")
print("   • Stuck orders may be test data, failed payments, or pipeline issues")
print("   • ACTION:")
print("     1. 🔍 INVESTIGATE: Check if created/paid orders ever transition → payment retry failures?")
print("     2. 📍 FLAG stuck orders separately for funnel analysis")
print("     3. 🎯 SEPARATE: Consider excluding from revenue metrics if undeliverable")
print("     4. 🔧 ROOT CAUSE: Check payment retry logic, payment processor callbacks")
```
